# SpaceX API Exploration

Understanding the data structure before processing everything.

## First requests

```python
# Fetch all launches
url = "https://api.spacexdata.com/v4/launches"
launches = requests.get(url).json()

print(f"Total launches: {len(launches)}")
print(f"\nFirst launch:")
print(json.dumps(launches[0], indent=2, default=str))
```

## Rocket relationship

```python
rocket_id = launches[0]["rocket"]
print(f"Rocket ID from first launch: {rocket_id}")

# Fetch rocket data
rocket = requests.get(
    f"https://api.spacexdata.com/v4/rockets/{rocket_id}"
).json()

print(f"\nRocket name: {rocket['name']}")
print(f"\nAll rocket properties:")
print(list(rocket.keys()))
```

## Core (booster) relationship

```python
cores_launch = launches[0]["cores"]
print(f"Cores in first launch: {cores_launch}")

# Get first core
core_id = cores_launch[0]["core"]
print(f"\nCore ID: {core_id}")

# Fetch core data
core = requests.get(
    f"https://api.spacexdata.com/v4/cores/{core_id}"
).json()

print(f"\nCore data:")
print(json.dumps(core, indent=2, default=str))
```

## Counting reuses

```python
launches_list = core.get("launches", [])
print(f"This core was reused in {len(launches_list)} launches")
print(f"Launch IDs: {launches_list})
```

## Available fields checklist

```python
print("Fields in Launch:")
launch = launches[0]
print(f"  - name: {launch.get('name')}")
print(f"  - flight_number: {launch.get('flight_number')}")
print(f"  - success: {launch.get('success')}")
print(f"  - date_utc: {launch.get('date_utc')}")
print(f"  - rocket (ID): {launch.get('rocket')}")
print(f"  - cores: {launch.get('cores')}")

print(f"\nFields in Rocket:")
print(f"  - name: {rocket.get('name')}")
print(f"  - id: {rocket.get('id')}")

print(f"\nFields in Core:")
print(f"  - id: {core.get('id')}")
print(f"  - launches: {len(core.get('launches', []))} (reuse count)")
```
