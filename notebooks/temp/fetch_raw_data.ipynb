# SpaceX API - Raw Data Extraction

Este notebook temporário faz o download dos dados brutos da API v4 da SpaceX e os salva na pasta `data/raw/`.

```python
import requests
import json
import os

# Garante que a pasta raw exista
raw_dir = '../../data/raw'
os.makedirs(raw_dir, exist_ok=True)
print(f"Diretório '{raw_dir}' pronto.")
```

```python
endpoints = {
    'launches': 'https://api.spacexdata.com/v4/launches',
    'cores': 'https://api.spacexdata.com/v4/cores',
    'rockets': 'https://api.spacexdata.com/v4/rockets'
}

for name, url in endpoints.items():
    print(f"Baixando dados de {name}...")
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        file_path = os.path.join(raw_dir, f'{name}_raw.json')
        
        with open(file_path, 'w', encoding='utf-8') as f:
            json.dump(data, f, indent=4, ensure_ascii=False)
            
        print(f"✅ Salvos {len(data)} registros em {file_path}\n")
    else:
        print(f"❌ Falha ao baixar {name}: Status {response.status_code}\n")
```
